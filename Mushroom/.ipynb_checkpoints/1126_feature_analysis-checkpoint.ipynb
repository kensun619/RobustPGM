## Attr 16 has only 1 value

```python
import numpy as np

lines = [line.rstrip('\n') for line in open('agaricus-lepiota.data')]

print(lines[0])
row = lines[0].split(',')
print(row)
print(lines[1])
row = lines[1].split(',')
print(row)
attr_list = []
for idx_attr in range(len(row)):
    attr_list.append([])
print(attr_list)
```

```python
data_num = []
for idx_lines in range(len(lines)-1):
    row = lines[idx_lines].split(',')
    entry = []
    for idx_attr in range(len(row)):
        if idx_attr != 16:
            if row[idx_attr] not in attr_list[idx_attr]:
                attr_list[idx_attr].append(row[idx_attr])
                entry.append(0)
            else:
                pos_attr = attr_list[idx_attr].index(row[idx_attr])
                entry.append(pos_attr)
    data_num.append(entry)
data_num = np.asarray(data_num)
print(data_num.shape)
```

```python
feature = data_num[:,1:]
label = data_num[:,0].reshape(-1, 1)
print("Feature matrix shape:", feature.shape)
print("Label vector shape:", label.shape)
print(np.unique(label))
```

```python
import pandas as pd

df_num = pd.DataFrame(data=data_num)
# print(df)
```

```python
n_col = data_num.shape[1]
max_attr_size = 13
ingd_ratio = []

for idx_col in range(n_col):
    col = data_num[:, idx_col]
    ingd = np.unique(col)
    col_ingd = [idx_col]
    for ins_ingd in ingd:
        col_ingd.append(list(col).count(ins_ingd))
    n_zero = max_attr_size-len(col_ingd)
    for idx_zero in range(n_zero):
        col_ingd.append(0)
    ingd_ratio.append(col_ingd)
ingd_ratio = np.array(ingd_ratio)
print(ingd_ratio)
```

```python
%matplotlib inline
import pandas as pd
import seaborn as sns
from matplotlib.colors import ListedColormap

col_name = ["Attr"]
for idx in range(max_attr_size-1):
    col_name.append("Value"+str(idx))
df = pd.DataFrame(columns=col_name, data=ingd_ratio)
# print(df)

```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.dpi"] = 300
df.set_index('Attr')\
  .reindex(df.set_index('Attr').sum().sort_values().index, axis=1)\
  .plot(kind='bar', stacked=True, title='Distributions in Each Attributes and Label',
        colormap=ListedColormap(sns.color_palette("muted", 24)),
        figsize=(12,6), legend=False)
```

```python
f, ax = plt.subplots(figsize=(10, 8))
corr = df_num.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool),
            cmap=sns.diverging_palette(240, 10, n=9, as_cmap=True),
            square=True, ax=ax)
plt.title('Correlations of columns')
```
